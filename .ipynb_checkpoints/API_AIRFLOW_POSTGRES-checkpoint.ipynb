#importing all the Requirements

```python
from airflow import DAG
from airflow.providers.https.hooks.http import HttpHook
from airflow.providers.postgres.hooks.postgres import postgresHook
from airflow.decorators import task
from airflow.utils.dates import days_ago
import requests
import JSON
```

```python
#Latitude and Longitude for the desired location (london in this case)
#setting_up connection
```

```python
LATITUDE = '51.5074'
LONGITUDE = '-0.1278'
POSTGRES_CONN_ID = 'Postgres_default'
API_CONN_ID = 'Open_Meteo_api'
```

#Creating Default arguments

```python
default_args = {
    'owner' : 'airflow'
    'start_date' : 'days_ago(1)'
}
```

#create a dag by using arguments

```python
with DAG (dag_id = 'weather_etl_pipeline',
         default_args = default_args,
         schedule_interval = '@daily',
         catchup = False) as dags:
    
    #creating a task for extraction process
    @task()
    def extract_weather_data():
        """"""Extract weather data from Open_meteo API using airflow connection """"""
        #use HTTP Hook to get the conn details from airflow connection
        
        http_hook = HttpHook(http_conn_id= API_CONN_ID,method = 'GET')
        
        #Build the API
        #https://api.open-meteo.com/VI/forecast?Latitude ={LATITUDE}& LONGITUDE = {LONGITUDE}&current_weather = true
        endpoint = f'/VI/forecast?Latitude = {LATITUDE}& LONGITUDE = {LONGITUDE} & current_weather = true'
        
        #make the req via the HTTp HOOK
        
        response = http.hook.run(endpoint)
        
        if reponse.statuscode == 200:
            return response.json()
        else
        raise exception(f"failed to fetch weather data:{response.statuscode}")
        
        #transformation
        
        @task()
        def_transform_weather_data(weather_data):
            """"""transform the weather data""""""
            
            current_weather = weather_data['current_weather']
            transformed_data = {
                'Latitude' : LATITUDE,
                'Longitude' : LONGITUDE,
                'temperature' : current_weather['temperature'],
                'windspeed' : current_weather['windspeed'],
                'winddirection' : current_weather['winddirection'],
                'weathercode' : current_weather['weathercode']
            }
            
        return transformed_data
    
    
    
    @task() def load_weather_data(transformed_data):
        """"""load transformed data into postgresql""""""
        
        pg.hook = postgresHook(POSTGRES_CONN_ID  = POSTGRES_CONN_ID)
        conn = pg.hook.get_conn()
        cursor = conn.cursor()
        
        
        #create table if doesn't exist
        cursor.execute(
        """"""CREATE TABLE IF NOT EXIST Weather_data(
        LATITUDE FLOAT
        LONGITUDE FLOAT
        TEMPERATURE FLOAT
        WINDSPEED FLOAT
        WINDDIRECTION FLOAT
        WEATHERCODE INT
        timestamp.TIMESTAMP DEFAULT CURRENT_TIMESTAMP);"""""")
        
        
        #INSERT transformed_data into the table
        
        cursor.execute(
            """"""INSERT INTO Weather_data(LATTITUDE,LONGITUDE,TEMPERATURE,WINDSPEED,WINDDIRECTION,
                                          WEATHERCODE
        VALUES(%S,%S,%S,%S,%S,%S)""""""),
        (
        transformed_data['LATITUDE'],
        transformed_data['LONGITUDE'],
        transformed_data['TEMPERATURE'],
        transformed_data['WINDSPEED'],
        transformed_data['WINDDIRECTION'],
        transformed_data['WEATHERCODE']))
        
        conn.commit()
        cursor.close()
```
